```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# csv_path = "learn_logs_25000.csv"
csv_path = "learn_logs_150000.csv"
df = pd.read_csv(csv_path)
df
```

```python
def plot_data(csv_path):
    """指定したcsvからグラフをプロット"""
    df = pd.read_csv(csv_path)

    fig = plt.figure()
    plt.subplots_adjust(wspace=0.4, hspace=0.6)

    ax1 = fig.add_subplot(211)  # 2行1列目の1番目
    ax1.plot(df["update"], df["reward"], linestyle="solid", label="reward")
    ax1.plot(df["update"], df["step"], linestyle="solid", label="step")
    ax1.legend(loc="best")
    ax1.set_title("reward by update")
    ax1.set_xlabel("update")
    ax1.set_ylabel("reward")

    ax2 = fig.add_subplot(212)  # 2行1列目の1番目
    # ax2.plot(df["update"], df["reward"], linestyle="solid", label="reward")
    ax2.plot(df["update"], df["r0"], linestyle="dashed", label="r0", alpha=0.6)
    ax2.plot(df["update"], df["r1"], linestyle="dashed", label="r1", alpha=0.6)
    ax2.legend(loc="best")
    ax2.set_title("step by update")
    ax2.set_xlabel("update")
    ax2.set_ylabel("step")

    plt.show()
```

## 通常報酬のみ

```python
csv_path = "learn_logs_150000_20221005215115.csv"
plot_data(csv_path)
```

## カスタム報酬

|報酬|説明|
|---|---|
|r0|通常の報酬|
|r1|カートが中央にいるほど高い報酬(0〜1)|

```python
# 中央にいるほど高報酬
# csv_path = "learn_logs_150000_20221005213128.csv"
# 右側(2-3)にいるほど高報酬
csv_path = "learn_logs_150000_20221005214653.csv"
plot_data(csv_path)
```
